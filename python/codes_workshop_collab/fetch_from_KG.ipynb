If you know the unique identifier of an instance, you can retrieve it directly:

```python
dataset_of_interest = DatasetVersion.from_id("4df8c324-af31-4cfc-a8af-a7ebf1d59fa1", kg_client)
dataset_of_interest.show()
```

Links between metadata in the Knowledge Graph are not followed automatically, to avoid unnecessary network traffic, but can be followed with the resolve() method:
KGProxy() objects represent links to other nodes in the graph.

```python
dataset_license = dataset_of_interest.accessibility.resolve(kg_client)
dataset_license.show()
```

```python
doi = dataset_of_interest.digital_identifier.resolve(kg_client)
doi.show()
```

```python
data_types = dataset_of_interest.data_types.resolve(kg_client)
data_types.show()
```

```python
techniques = [tech.resolve(kg_client) for tech in dataset_of_interest.techniques]
for tech in techniques:
    print(tech.name)
```

We can also follow reverse links, i.e. where the property of another object points to the dataset version. Reverse properties are not shown by default by obj.show(), but you can see them by passing the include_empty_properties option:

```python
dataset_of_interest.show(include_empty_properties=True)
```

Reverse properties are represented by a KGQuery object. These can also be resolved:

```python
example_dataset = dataset_of_interest.is_version_of.resolve(kg_client)
example_dataset.show()
```

Any combination of forward and reverse links can be followed.

Following links can even be done at the same time as retrieving the base node. This request includes following a "reverse" link ("is_version_of") followed by a "forward" link ("authors")

```python
#maya's test dataset version
dataset_version = DatasetVersion.from_id(
    '724d4af0-fe28-4032-8837-120b0d64a81c',
    kg_client,
    scope="in progress",
    follow_links={
        "data_types": {},
        "is_version_of": {
        "authors": {}
        }
    }
)
dataset_version.show()
```

# Writing to the Knowledge Graph

```python
from fairgraph.openminds.controlledterms import SemanticDataType
data_types_list = SemanticDataType.list(kg_client, size=100, from_index=0, api='auto', scope='released', resolved=False, space=None)
data_types_names = [item.name for item in data_types_list]
```

```python
print('Please select data type:')
print('---')
c = 0
for d in data_types_names:
    c += 1
    print(str(c) + ". " + d)
print('---')
selected = input('Your choice: ')
index = int(selected)-1
print('')
print(data_types_names[index])
```

```python
dataset_test_maya.data_types = SemanticDataType.by_name(data_types_names[index], kg_client)
```
