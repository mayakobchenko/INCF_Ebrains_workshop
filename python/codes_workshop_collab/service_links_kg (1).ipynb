```python
from fairgraph.openminds.core import ServiceLink
```

```python
#returns 10 nodes starting with the 15th
service_link_list = ServiceLink.list(kg_client, from_index=15, size=2)
```

```python
links_siibra = omcore.ServiceLink.list(
    kg_client,
    from_index=1, size=2,
    service__name="siibra-explorer",
    follow_links={"service": {}}  
)
print(f"Found {len(links_siibra)} links")

links_siibra[0].service.resolve(kg_client).name
```

```python
linked_from = links_siibra[0].service.resolve(kg_client).linked_from
```

```python
for i in linked_from:
    print(i.resolve(kg_client))
```

```python
types_list = links_siibra[0].data_location.resolve(kg_client).file_repository.resolve(kg_client).content_type_patterns
```

```python
for i in types_list:
    print(i.resolve(kg_client).lookup_label)
```

```python
for i in types_list:
    print(i.resolve(kg_client).show())
```
