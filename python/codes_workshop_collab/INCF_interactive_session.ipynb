If you know the unique identifier of an instance, you can retrieve it directly:

```python
dataset_of_interest = DatasetVersion.from_id("4df8c324-af31-4cfc-a8af-a7ebf1d59fa1", kg_client)
dataset_of_interest.show()
```

Links between metadata in the Knowledge Graph are not followed automatically, to avoid unnecessary network traffic, but can be followed with the resolve() method:
KGProxy() objects represent links to other nodes in the graph.

```python
access = dataset_of_interest.accessibility.resolve(kg_client)
access.show()
```

```python
doi = dataset_of_interest.digital_identifier.resolve(kg_client)
doi.show()
```

```python
data_types = dataset_of_interest.data_types.resolve(kg_client)
data_types.show()
```

```python
techniques = [tech.resolve(kg_client) for tech in dataset_of_interest.techniques]
for tech in techniques:
    print(tech.name)
```

We can also follow reverse links, i.e. where the property of another object points to the dataset version. Reverse properties are not shown by default by obj.show(), but you can see them by passing the include_empty_properties option:

```python
dataset_of_interest.show(include_empty_properties=True)
```

Reverse properties are represented by a KGQuery object. These can also be resolved:

```python
example_dataset = dataset_of_interest.is_version_of.resolve(kg_client)
example_dataset.show()
```

Any combination of forward and reverse links can be followed.

Following links can even be done at the same time as retrieving the base node. This request includes following a "reverse" link ("is_version_of") followed by a "forward" link ("authors")

```python
#maya's test dataset version
dataset_version = DatasetVersion.from_id(
    '724d4af0-fe28-4032-8837-120b0d64a81c',
    kg_client,
    scope="in progress",
    follow_links={
        "data_types": {},
        "is_version_of": {
        "authors": {}
        }
    }
)
dataset_version.show()
```

Retrieving a dataset based on its name

```python
dataset_versions = DatasetVersion.list(kg_client, full_name="Excitability profile of CA1 pyramidal neurons in APPPS1 Alzheimer disease mice and control littermates")
print(f"Found {len(dataset_versions)} dataset(s)")
```

we don't need the complete name, fairgraph will also search for fragments of the name

```python
dataset_versions = DatasetVersion.list(kg_client, full_name="Alzheimer")
print(f"Found {len(dataset_versions)} dataset(s)")
dataset_versions[0].full_name
```

```python
from fairgraph.openminds.core import Dataset
alz_ds = Dataset.list(kg_client, full_name="Alzheimer")
print(f"Found {len(alz_ds)} dataset(s)")
```

```python
#alias is short name of a dataset version
alzheimers_datasets = DatasetVersion.list(
    kg_client,
    is_version_of__full_name="Alzheimer",
    is_version_of__description="Alzheimer"
)
print(f"Found {len(alzheimers_datasets)} dataset(s)")
for alz_dsv in alzheimers_datasets:
    print(alz_dsv.alias)
```

```python
Alzheimer_count = DatasetVersion.count(kg_client, is_version_of__full_name="Alzheimer",
    is_version_of__description="Alzheimer")
Alzheimer_count
```

Controlled terms

Many properties in openMINDS, like "study_targets" for example, use controlled vocabularies, which means that the property is a link to a node, which in turn contains both the term itself and other information about the term, such as a definition.

```python
from fairgraph.openminds import controlled_terms 
help(controlled_terms)
```

```python
dsv_properties = DatasetVersion.properties
for pr in dsv_properties:
    print(pr.name)
```

```python
DatasetVersion.property_names
```

```python
DatasetVersion._property_lookup["study_targets"].types
```

```python
controlled_terms.Disease.property_names
```

```python
possible_terms = controlled_terms.Disease.list(kg_client, name="Alzheimer's disease")
print(f"Found {len(possible_terms)} term(s)")

alzheimers = possible_terms[0]
alzheimers.show()
```
